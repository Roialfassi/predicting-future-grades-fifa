## Test Wikipedia Data

```python
import wikipedia
import re

def get_player_stats(last_name, team, age):
    # Construct the Wikipedia page title based on the player's last name and team
    page_title = f"{last_name} {team} season {age}-{age+1}"

    try:
        # Load the Wikipedia page
        page = wikipedia.page(page_title)

        # Extract the player's statistics using regular expressions
        appearances = re.search(r"Appearances:\s*(\d+)", page.content).group(1)
        assists = re.search(r"Assists:\s*(\d+)", page.content).group(1)
        goals = re.search(r"Goals:\s*(\d+)", page.content).group(1)

        # Return the statistics as a tuple
        return int(appearances), int(assists), int(goals)
    except wikipedia.exceptions.PageError:
        # Return None if the page doesn't exist
        return None
    except (AttributeError, ValueError):
        # Return None if the statistics couldn't be extracted
        return None
```

```python
appearances, assists, goals = get_player_stats("Ronaldo", "Real Madrid", 30)
print(f"Appearances: {appearances}, Assists: {assists}, Goals: {goals}")
```
